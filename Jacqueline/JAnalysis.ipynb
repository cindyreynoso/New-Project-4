## Do imports, reads in csv, create region dfs

```python
# the imports
import pandas as pd
from regions import NorthAmerica, CentralAmerica, SouthAmerica, Europe, MiddleEast, SubSaharanAfrica, SouthAsia, SouthEastAsia, EastAsia, Oceania
```

```python
# read in the data
mhddd_df = pd.read_csv("../Mental health Depression disorder Data.csv")
mhddd_df.head()
```

```python
#NorthAmerica, CentralAmerica, SouthAmerica, Europe, MiddleEast, SubSaharanAfrica, SouthAsia, SouthEastAsia, EastAsia, Oceania
NA_df = mhddd_df.loc[mhddd_df["country"].isin(NorthAmerica)]

CA_df = mhddd_df.loc[mhddd_df["country"].isin(CentralAmerica)]

SA_df = mhddd_df.loc[mhddd_df["country"].isin(SouthAmerica)]

EU_df = mhddd_df.loc[mhddd_df["country"].isin(Europe)]

ME_df = mhddd_df.loc[mhddd_df["country"].isin(MiddleEast)]

SSA_df = mhddd_df.loc[mhddd_df["country"].isin(SubSaharanAfrica)]

SAsia_df = mhddd_df.loc[mhddd_df["country"].isin(SouthAsia)]

SEAsia_df = mhddd_df.loc[mhddd_df["country"].isin(SouthEastAsia)]

EAsia_df = mhddd_df.loc[mhddd_df["country"].isin(EastAsia)]

OC_df = mhddd_df.loc[mhddd_df["country"].isin(Oceania)]
```

## Do certain regions have higher rates of mental health percentages?


```python
#Do certain regions have higher rates of mental health percentages? (e.g. in some countries it might be more socially accepted)

# What do I want to do with this question?
# Find average diagnosis rates of each disorder in each region
# Compare that overall -> which disorder most prevalent overall
# Then compare them on an individual basis -> compare anxiety across regions (bar chart), compare alchoholism across regions

# Create a dataframe with averages across all regions
columns = ["schizophrenia", "bipolar_disorder", "eating_disorders", "anxiety_disorders", "drug_use_disorders", "depression", "alcohol_use_disorders"]

Q1_df = pd.DataFrame(NA_df[columns].mean())
Q1_df = Q1_df.rename({0 : "NorthAmerica"}, axis = 1)

Q1_df["CentralAmerica"] = (CA_df[columns].mean())
Q1_df["SouthAmerica"] = (SA_df[columns].mean())
Q1_df["Europe"] = (EU_df[columns].mean())
Q1_df["MiddleEast"] = (ME_df[columns].mean())
Q1_df["SubSaharanAfrica"] = (SSA_df[columns].mean())
Q1_df["SouthAsia"] = (SAsia_df[columns].mean())
Q1_df["SouthEastAsia"] = (SEAsia_df[columns].mean())
Q1_df["EastAsia"] = (EAsia_df[columns].mean())
Q1_df["Oceania"] = (OC_df[columns].mean())
```

```python
Q1a_df = Q1_df.transpose()
Q1a_df
```

```python
# Include overall average, max, and min
Q1a_df["Average"] = Q1a_df.mean(axis = 1)
Q1a_df["High"] = Q1a_df.max(axis = 1)
Q1a_df["Low"] = Q1a_df.min(axis = 1)
Q1a_df
```

```python
fig1 = Q1a_df[["Average", "High", "Low"]].plot(kind = "bar")
```

```python
fig1a = Q1a_df[["schizophrenia", "bipolar_disorder", "eating_disorders", "anxiety_disorders", 
                "drug_use_disorders", "depression","alcohol_use_disorders"]].plot(kind = "line", use_index=False, )
fig1a.set_xlabel("Countries")
fig1a.set_ylabel("Diagnosis")
```

```python
fig1b = Q1a_df[["anxiety_disorders", "depression", "eating_disorders", "schizophrenia"]].plot(kind = "bar", color = ["red", "brown", "green", "blue"], alpha = 0.7)
```
